## Unicode

```python
x = u'abc'
print(x)
print(type(x))
```

```python
x = u'Элеонора Михайловна'
print(x, type(x))
y = x.encode('utf-8')
print(y, type(y))
z = y.decode('utf-8')
print(z, type(z))
q = y.decode('cp1251')
print(q, type(q))
```

```python
#явное приведение типов не сработает!
print(str(x))
```

```python
#возьмем срез от строки, полученной из unicode
print(x[1:])
print(y[1:])
#размер строки тоже разный
print(len(x))
print(len(y))
```

```python
y = u'Иван Иванович'.encode('utf-8')
print(y.decode('utf-8'))
print(y.decode('cp1251'))
```
