## Оптимизация

```python
from scipy import optimize
```

```python
def f(x):
    return (x[0] - 3.2) ** 2 + (x[1] - 0.1) ** 2 + 3

print f([3.2, 0.1])
```

```python
x_min = optimize.minimize(f, [5, 5])
print x_min
```

```python
x_min = optimize.minimize(f, [5, 5], method='Nelder-Mead')
print x_min
```

```python
print x_min.x
```

## Линейная алгебра

```python
from scipy import linalg
```

```python
a = np.array([[3, 2, 0], [1, -1, 0], [0, 5, 1]])
b = np.array([2, 4, -1])

x = linalg.solve(a, b)
print x
```

```python
print np.dot(a, x) # проверили результат
```

```python
X = np.random.randn(4, 3)
U, D, V = linalg.svd(X)
print U.shape, D.shape, V.shape
print type(U), type(D), type(V)
print V
```

## Все вместе

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.arange(0, 10, 2)
y = np.exp(-x/3.0) + np.random.randn(len(x)) * 0.05

print x[:5]
print y[:5]
```

```python
f = interpolate.interp1d(x, y, kind='quadratic')
xnew = np.arange(0, 8, 0.1)
ynew = f(xnew)
```

```python
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

```python
x = np.arange(0, 10, 2)
y = np.exp(-x/3.0)

print x[:5]
print y[:5]
```

```python
f = interpolate.interp1d(x, y, kind='linear')
xnew = np.arange(0, 8, 0.1)
ynew = f(xnew)
```

```python
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```
